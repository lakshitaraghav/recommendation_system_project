```python
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):  # "obj" string becomes list with the help of this function
        L.append(i['name'])
    return L
```

```python
import ast #module 
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
def main_cast(obj): # taking only first 6 characters of the film
    L =[]
    counter = 0
    for i in ast.literal_eval(obj):  
        if counter !=3:
            L.append(i['name'])
            counter +=1
        else:
            break
        return L
```

```python
movies['cast'].apply(main_cast)
```

```python
movies['cast'] = movies['cast'].apply(main_cast)
```

```python
movies.head()
```

```python
def movie_director(obj): # finding director of the film from list
    L=[]
    for i in ast.literal_eval(obj): 
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew']=movies['crew'].apply(movie_director)
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)    
```

```python
movies['tags']= movies['tags'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True, key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    index=movies[movies['title']==movie].index[0]
    distances = similarity[index]
    movies_list= sorted(list(enumerate(distances)),reverse=True,key = lambda x:x[1])[1:6]
                       
    for i in movies_list:
        print(movies.iloc[i[0]].title)
        
```

```python
recommend('Spectre')
```

```python
recommend('The Dark Knight Rises')
```
